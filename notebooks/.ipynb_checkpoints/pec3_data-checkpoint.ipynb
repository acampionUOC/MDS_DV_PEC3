# Dataset generation for visualization

```python
import pandas as pd
import numpy as np
```

## Información partidos políticos con reprentación parlamentaria en 2021 o 2024

```python
# dataframe for political groups
df_gp = pd.DataFrame(dict(name=['PSC', 'CAT-JUNTS+', 'ERC', 'PP', 'VOX', 'COMUNS SUMAR',
                                'CUP - DT', 'ALIANÇA.CAT', 'Cs'],
                          rep_24 = [42, 35, 20, 15, 11, 6,
                                   4, 2, 0],
                          rep_21 = [33, 32, 33, 3, 11, 8,
                                   9, 0, 6],
                          name_21=['PSC', 'JxCat', 'ERC', 'PP', 'VOX', 'ECP-PEC',
                                  'CUP-G', 'ALIANÇA.CAT', 'Cs'],
                         ind=[0, 1, 1, 0, 0, 0, 
                              1, 1, 0],
                    pos = ['CI', 'C', 'CI-I', 'CD-D', 'D-ED', 'CI-I',
                          'I-EI', 'ED', 'C-CD'],
                    pos_num=[1.5, 2, 1.25, 2.75, 3.5, 1.25,
                            0.5, 4, 2.25]))

# resultados provinciales

df_gp['Barcelona_24'] = [28, 18, 12, 11, 7, 6,
                                   3, 0, 0]
df_gp['Barcelona_21'] = [23, 16, 19, 3, 7, 7,
                                   5, 0, 5]
df_gp['Girona_24'] = [4, 7, 2, 1, 1, 0,
                                   1, 1, 0]
df_gp['Girona_21'] = [3, 7, 4, 0, 1, 0,
                                   2, 0, 0]
df_gp['Tarragona_24'] = [6, 5, 3, 2, 2, 0,
                                   0, 0, 0]
df_gp['Tarragona_21'] = [4, 4, 5, 0, 2, 1,
                                   1, 0, 1]
df_gp['Lleida_24'] = [4, 5, 3, 1, 1, 0,
                                   0, 1, 0]
df_gp['Lleida_21'] = [3, 5, 5, 0, 1, 0,
                                   1, 0, 0]
df_gp['pos_num'] /= 4
df_gp
```

## Información geográfica municipios
https://centrodedescargas.cnig.es/CentroDescargas/catalogo.do?Serie=CAANE <br>
Nomenclátor Geográfico de Municipios y Entidades de Población

```python
# https://centrodedescargas.cnig.es/CentroDescargas/catalogo.do?Serie=CAANE
# Nomenclátor Geográfico de Municipios y Entidades de Población
df_mu = pd.read_csv(r'..\input\MUNICIPIOS.csv', sep=';', encoding='mbcs')
```

```python
# filter catalunya
df_mu = df_mu[df_mu['COD_PROV'].isin([17, 25, 43, 8])]
df_mu['COD_INE'] = (df_mu['COD_INE'] / 1000000).astype(int)
df_mu
```

## Carga y procesado resultados 12M 2024

```python
# https://resultados.eleccionsparlament.cat/es/inicio
# resultados provisionales 12M 2024
df_vot = pd.read_csv(r'..\input\epc_2024.csv', sep=';', encoding='utf-8-sig')
```

```python
# Filter municipios y provincias
df_vot = df_vot[df_vot['Nivell'].isin(['MU', 'CI'])]
df_vot['COD_INE'] = (df_vot['Codi circumscripció'] * 1000 + df_vot['Codi municipi'].fillna(0)).astype(int)
# add year
df_vot['any'] = 2024
```

```python
# añade en Nom municipi el nombre de la PROVICIA y COD_INE de la capital para el Nivel CI
provs_code = np.unique(df_vot['Codi circumscripció'])
for p in provs_code:
    # find name and cod_ine
    name_prov = df_mu[df_mu['COD_PROV'] == p]['PROVINCIA'].iloc[0]
    cod_ine = df_vot[df_vot['Nom municipi'] == name_prov]['COD_INE'].iloc[0]
    COD_INE_cap = df_vot[df_vot['Nom municipi'] == name_prov]['COD_INE'].iloc[0]
    
    # update in df_vot
    f = (df_vot['Nivell'] == 'CI') & (df_vot['Codi circumscripció'] == p)
    df_vot.loc[f, 'Nom municipi'] = name_prov
    df_vot.loc[f, 'COD_INE'] = int(COD_INE_cap)
    
```

```python
# calcula ideología promedio - considera solo partidos con representación alguno de los años
# añade columna con numero de votos por partido con representacion según año 2024
# añade columna con votos nacionalistas

# 2024
df_vot['ideologia_avg'] = 0
df_vot['nacionalismo_avg'] = 0
df_vot['nacionalismo_vots'] = 0
df_vot['vots_count'] = 0
for row in df_gp.itertuples():
    vots_col = ' Vots {} '.format(row.name)
    if vots_col in df_vot.columns:
        df_vot['ideologia_avg'] += df_vot[vots_col] * row.pos_num
        df_vot['nacionalismo_avg'] += df_vot[vots_col] * row.ind
        df_vot['nacionalismo_vots'] += df_vot[vots_col] * row.ind
        df_vot['vots_count'] += df_vot[vots_col]
        df_vot['{}_vots'.format(row.name)] = df_vot[vots_col]
    else:
        df_vot['{}_vots'.format(row.name)] = 0

df_vot['ideologia_avg'] /= df_vot['vots_count']
df_vot['nacionalismo_avg'] /= df_vot['vots_count']
```

```python
# filtramos columnas 
df_vot['Cens'] = df_vot['Cens electoral']
f_cols = ['Nivell', 'Cens', 'any', 'ideologia_avg', 'nacionalismo_avg', 'nacionalismo_vots', 'vots_count', 'PSC_vots',
       'CAT-JUNTS+_vots', 'ERC_vots', 'PP_vots', 'VOX_vots', 'COD_INE',
       'COMUNS SUMAR_vots', 'CUP - DT_vots', 'ALIANÇA.CAT_vots', 'Cs_vots']

df_vot = df_vot[f_cols]
df_vot
```

## Carga y procesado resultados 14F 2021

```python
df_vot_21 = pd.read_csv(r'..\input\A20211-Columnes-MU.csv', sep=';', encoding='utf-8-sig')

df_vot_21['COD_INE'] = df_vot_21['Codi Municipi'].astype(int)
# add year
df_vot_21['any'] = 2021
df_vot_21['Nivell'] = 'MU'
# filter Residents Absents
df_vot_21 = df_vot_21[df_vot_21['Nom Municipi'] != 'Residents Absents']
```

```python
# agrupa resultados por provincia, actualiza valores en las columnas que se utilizaran en  la visualización
for p in provs_code:
    # find name and cod_ine
    name_prov = df_mu[df_mu['COD_PROV'] == p]['PROVINCIA'].iloc[0]
    COD_INE_cap = df_vot_21[df_vot_21['Nom Municipi'] == name_prov]['COD_INE'].iloc[0]
    
    # generamos nuevo row
    df_vot_21 = df_vot_21._append(df_vot_21.iloc[0],ignore_index=True)
    
    df_vot_21.loc[df_vot_21.index[-1], 'Nivell'] = 'CI'
    df_vot_21.loc[df_vot_21.index[-1], 'COD_INE'] = COD_INE_cap
    df_vot_21.loc[df_vot_21.index[-1], 'Nom Municipi'] = name_prov
    df_vot_21.loc[df_vot_21.index[-1], 'Codi Província'] = p
    
    
    # filter in provincia
    f_in_prov = df_vot_21['Codi Província'] == int(p)
    cols_to_sum = ['Cens'] +  ['{} Vots'.format(x) for x in df_gp.name_21 if '{} Vots'.format(x) in df_vot_21.columns]
    for c in cols_to_sum:
        # primero se pone a 0 para evitar sumar en la suma global
        df_vot_21.loc[df_vot_21.index[-1], c] = 0
        df_vot_21.loc[df_vot_21.index[-1], c] = df_vot_21.loc[f_in_prov, c].sum()
    
   
```

```python
# mismo procesado para 2021
df_vot_21['ideologia_avg'] = 0
df_vot_21['nacionalismo_avg'] = 0
df_vot_21['nacionalismo_vots'] = 0
df_vot_21['vots_count'] = 0
for row in df_gp.itertuples():
    vots_col = '{} Vots'.format(row.name_21)
    if vots_col in df_vot_21.columns:
        df_vot_21['ideologia_avg'] += df_vot_21[vots_col] * row.pos_num
        df_vot_21['nacionalismo_avg'] += df_vot_21[vots_col] * row.ind
        df_vot_21['nacionalismo_vots'] += df_vot_21[vots_col] * row.ind
        df_vot_21['vots_count'] += df_vot_21[vots_col]
        df_vot_21['{}_vots'.format(row.name)] = df_vot_21[vots_col]
    else:
        df_vot_21['{}_vots'.format(row.name)] = 0

df_vot_21['ideologia_avg'] /= df_vot_21['vots_count']
df_vot_21['nacionalismo_avg'] /= df_vot_21['vots_count']

```

```python
# filtramos columnas 
df_vot_21 = df_vot_21[f_cols]
df_vot_21
```

## añade calculo variación 2021 a 2024

```python
df_vot_21 = df_vot_21.sort_values(by=['Nivell', 'COD_INE']).reset_index(drop=True)
df_vot = df_vot.sort_values(by=['Nivell', 'COD_INE']).reset_index(drop=True)
```

```python
df_vot_21['ideologia_avg_delta'] = df_vot['ideologia_avg'] - df_vot_21['ideologia_avg']
df_vot['ideologia_avg_delta'] = df_vot['ideologia_avg'] - df_vot_21['ideologia_avg']
df_vot_21['nacionalismo_avg_delta'] = df_vot['nacionalismo_avg'] - df_vot_21['nacionalismo_avg']
df_vot['nacionalismo_avg_delta'] = df_vot['nacionalismo_avg'] - df_vot_21['nacionalismo_avg']
```

## Fusión resultados 2021 y 2024 e integración información geográfica

```python
# concat dataFrames
df = pd.concat([df_vot, df_vot_21], ignore_index=True)
```

```python
# merge with Longitude and latitude
df = df.merge(df_mu[['COD_INE', 'NOMBRE_ACTUAL', 'PROVINCIA', 'LONGITUD_ETRS89', 'LATITUD_ETRS89']],
             left_on='COD_INE', right_on='COD_INE')
```

```python
# añade censo_logarítmico
df['Cens_log'] = np.log(df['Cens'])
df
```

```python
df.to_csv(r'..\output\test_map.csv')
```
